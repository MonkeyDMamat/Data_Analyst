```python
#Load Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv("C:/MK Semester 4/Data Mining/Mlbb_Heroes.csv")
```

```python
# Histogram
plt.hist(df["Lane"], bins=10)
plt.title("Histogram of Lane")
plt.xlabel("Lane")
plt.ylabel("Frequency")
plt.show()

# Jenis visualisasi data yang digunakan untuk menampilkan distribusi frekuensi dari suatu variabel numerik.
# Histogram digunakan pada Jupyter Notebook karena dapat memberikan informasi yang berguna dalam analisis data seperti melihat pola distribusi data, menemukan nilai tengah atau rata-rata, dan mengidentifikasi outlier.
```

```python
# ScatterPlot
plt.scatter(df["Lane"], df["Hp"])
plt.title("Scatter Plot of Lane vs Hp")
plt.xlabel("Lane")
plt.ylabel("Hp")
plt.show()

# Jenis visualisasi data yang digunakan untuk menampilkan hubungan antara dua variabel numerik dalam bentuk titik yang tersebar di bidang kartesian.
# Scatter plot digunakan pada Jupyter Notebook karena dapat memberikan informasi yang berguna dalam analisis data seperti melihat hubungan antara dua variabel, menemukan pola atau tren, dan memperlihatkan outlier.
```

```python
# BarPlot
df["Primary_Role"].value_counts().plot(kind="bar")
plt.title("Bar Plot of Primary_Role")
plt.xlabel("Primary_Role")
plt.ylabel("Frequency")
plt.show()

# Jenis visualisasi data yang digunakan untuk menampilkan frekuensi atau proporsi data kategori dalam bentuk batang vertikal atau horizontal.
# Bar plot digunakan pada Jupyter Notebook karena dapat memberikan informasi yang berguna dalam analisis data seperti melihat perbandingan antara beberapa kategori data, menunjukkan tren, dan memperlihatkan perbedaan proporsi atau frekuensi antara kategori data.
```

```python
# BoxPlot
df.boxplot(column=["Esport_Wins", "Esport_Loss"])
plt.title('Box Plot of Esport_Wins and Esport_Loss')
plt.ylabel('Value')
plt.show()

# Salah satu jenis visualisasi data yang digunakan untuk menampilkan distribusi data numerik dalam bentuk diagram yang menunjukkan nilai-nilai ekstrim, kuartil, median, dan outlier.
# Box plot dapat memberikan informasi yang berguna dalam analisis statistik seperti melihat persebaran data, menemukan pencilan (outlier), melihat distribusi data, serta membandingkan distribusi antara beberapa grup atau kategori data.
```

```python
# HeatMap
sns.heatmap(df.corr(), annot=True)
plt.title('Heatmap of Correlation')
plt.show()

# Sebuah visualisasi yang digunakan untuk menampilkan data dalam bentuk matriks warna yang memperlihatkan korelasi antara variabel dalam suatu dataset.
```

```python
# Visualize the distribution of hero roles
plt.figure(figsize=(8,6))
sns.countplot(x="Primary_Role", data=df)
plt.title("Distribution of Hero Roles")
plt.show()

# Dengan menggunakan sns.countplot() untuk menghitung jumlah hero untuk tiap role.
# Setiap visualisasi dilengkapi dengan judul (plt.title()), label sumbu-x (plt.xlabel()), dan label sumbu-y (plt.ylabel()).
```

```python
# Visualize the relationship between hero win rate and lose rate
plt.figure(figsize=(8,6))
sns.scatterplot(x="Esport_Wins", y="Esport_Loss", data=df)
plt.title("Esport_Wins vs Esport_Loss")
plt.xlabel("Esport_Wins")
plt.ylabel("Esport_Loss")
plt.show()

# Dengan menggunakan sns.scatterplot() untuk menampilkan hubungan antara win rate dan lose rate dari hero yang digunakan di beberapa pertandingan. 
# Setiap visualisasi dilengkapi dengan judul (plt.title()), label sumbu-x (plt.xlabel()), dan label sumbu-y (plt.ylabel()).
```

```python
# Visualize the distribution of hero Mag_Damage and Mag_Defence
plt.figure(figsize=(12,8))
sns.violinplot(x="Mag_Damage", y="Mag_Defence", data=df)
plt.title("Mag_Damage by Mag_Defence")
plt.xlabel("Mag_Damage")
plt.ylabel("Mag_Defence")
plt.show()

# Dengan menggunakan sns.violinplot() untuk menampilkan distribusi tingkat Damage Magical dan Defence Magical Hero.
# Setiap visualisasi dilengkapi dengan judul (plt.title()), label sumbu-x (plt.xlabel()), dan label sumbu-y (plt.ylabel()).
```

```python
# Mengisi Missing Value
df["Mag_Defence"].fillna(df["Mag_Defence"].mean(), inplace=True)

#Missing value (nilai kosong atau null) dapat mengganggu proses analisis data dan dapat menghasilkan kesalahan dalam analisis. Oleh karena itu, penting untuk mengisi atau menghapus missing value sebelum melakukan analisis data. 
```

```python
# Mengubah Format Data
df["Mov_Speed"] = df["Mov_Speed"].astype('int64')

# proses untuk mengubah tipe data suatu kolom pada dataset menjadi tipe data yang berbeda. Format data dapat berupa jenis data yang berbeda seperti float, integer, string, datetime, atau kategori.
```

```python
# Membersihkan Data yang Tidak Valid
df = df[df["Mag_Defence"] >= 0]

# proses penting dalam analisis data. Data yang tidak valid dapat menyebabkan bias atau ketidakakuratan dalam hasil analisis.
```

```python
# Melakukan Standardisasi pada Data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[["Mana", "Mana_Regen"]] = scaler.fit_transform(df[["Mana", "Mana_Regen"]])

# sebuah proses normalisasi data yang digunakan untuk membuat setiap variabel dalam suatu dataset memiliki skala yang sama, sehingga memudahkan dalam membandingkan dan menginterpretasikan nilai-nilai variabel. Proses standardisasi dilakukan dengan mengurangi nilai rata-rata variabel dan membagi dengan standar deviasi dari variabel tersebut.
```
