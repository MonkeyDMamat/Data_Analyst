```python
# Load Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Data
df = pd.read_csv("C:/MK Semester 4/Data Mining/Mlbb_Heroes.csv")
```

```python
# Mengecek dataset
df.head()
```

```python
# Memeriksa informasi dari dataset
df.info()
```

```python
# Memeriksa apakah terdapat nilai yang hilang
df.isnull().sum()
```

```python
# Menghilangkan kolom yang tidak diperlukan
df = df.drop(["Title", "Voice_Line", "Release_Date"], axis=1)
```

```python
# Mengganti nama kolom (Yang tadi nya huruf kapital menjadi tidak)
df = df.rename(columns={"Hp":"hp", "Lane":"lane", "Mana":"mana"})
```

```python
# Menghapus data duplikat
df = df.drop_duplicates()
```

```python
# Mengisi nilai yang hilang
df["Mag_Damage"].fillna("None", inplace=True)
```

```python
# Menghilangkan spasi pada nama kolom
df.columns = df.columns.str.replace(" ", "")
```
