```python
# Load Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data
df = pd.read_csv("C:/MK Semester 4/Data Mining/Mlbb_Heroes.csv")
```

```python
# Menghitung Statistik Deskriptif
df.describe()

# Teknik analisis statistik yang digunakan untuk merangkum dan menggambarkan karakteristik atau sifat dari suatu dataset. Dengan mengetahui statistik deskriptif dari suatu variabel, kita dapat memahami lebih baik tentang karakteristik atau sifat dari variabel tersebut.
```

```python
# Menghitung Korelasi
df.corr()

# Pengukuran hubungan antara dua variabel dalam sebuah dataset. Korelasi dapat digunakan untuk mengukur kekuatan dan arah hubungan antara dua variabel, apakah positif atau negatif.  Dengan mengetahui korelasi antara dua variabel tersebut, kita dapat memahami lebih baik tentang hubungan antara kedua variabel tersebut dan membuat keputusan yang lebih baik.
```

```python
# Membuat Histogram
plt.hist(df["Mov_Speed"], bins=10)
plt.show()

# Histogram digunakan pada Jupyter Notebook karena dapat memberikan informasi yang berguna dalam analisis data seperti melihat pola distribusi data, menemukan nilai tengah atau rata-rata, dan mengidentifikasi outlier.
```

```python
# Membuat Scatter Plot
plt.scatter(df["Phy_Damage"], df["Phy_Defence"])
plt.show()

# Scatter plot digunakan pada Jupyter Notebook karena dapat memberikan informasi yang berguna dalam analisis data seperti melihat hubungan antara dua variabel, menemukan pola atau tren, dan memperlihatkan outlier.
```

```python
# Melakukan Uji Hipotesis
from scipy.stats import ttest_ind

group1 = df[df["Name"] == "Assassin"]["Phy_Damage"]
group2 = df[df["Name"] == "Mage"]["Phy_Damage"]

ttest_ind(group1, group2)

# Proses pengambilan keputusan statistik yang didasarkan pada data sampel dengan tujuan untuk membuat kesimpulan tentang populasi.
```

```python
# Melakukan Regresi Linear
from sklearn.linear_model import LinearRegression

x = df[["Esport_Wins"]]
y = df[["Esport_Loss"]]

model = LinearRegression()
model.fit(x, y)

plt.scatter(x, y)
plt.plot(x, model.predict(x), color='red')
plt.show()

#metode analisis statistik untuk menentukan hubungan antara variabel independen (X) dan variabel dependen (Y) dengan asumsi bahwa hubungan tersebut dapat dijelaskan oleh sebuah garis lurus. Pada Jupyter Notebook, regresi linear dapat dilakukan dengan menggunakan library statistik seperti SciPy dan scikit-learn.
```

```python
# Histplot (fungsi tingkat sumbu untuk histogram)
sns.histplot(df["Hp_Regen"])
plt.title("Distribution of Hp_Regen")
plt.show()

# Untuk melihat pola distribusi data secara visual dan membantu untuk menentukan jenis analisis statistik yang tepat untuk digunakan. Histogram plot dapat membantu dalam memahami beberapa karakteristik data seperti apakah data tersebut mengikuti distribusi normal, apakah terdapat outlier, dan apakah data tersebut memiliki skewness atau tidak.
```
